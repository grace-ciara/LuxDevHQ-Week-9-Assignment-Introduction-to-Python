## Working with JSON Data from GitHub
* Using realestate_data dataset

```python
import requests
```

```python
url = "https://raw.githubusercontent.com/grace-ciara/LuxDevHQ-Week-9-Assignment-Introduction-to-Python/refs/heads/main/realestate_data.json"
```

```python
estate_data = requests.get(url)
```

```python
estate_data.status_code
```

```python
content = estate_data.json()
content
```

```python
type(content)
```

```python
data_10 = content[:10]
len(content[:10])
```

```python
import pandas
```

```python
df = pandas.DataFrame(data_10)
```

```python
df.head(10)
```

```python
df.to_csv("realestatedata.csv")
```
