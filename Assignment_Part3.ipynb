## Working with DummyJSON API Endpoints
Using the following DummyJSON endpoints:

1. Products endpoint
  https://dummyjson.com/products

2. Carts endpoint
  https://dummyjson.com/carts


#### 1. Products endpoint

```python
import requests
```

```python
url = "https://dummyjson.com/products"
```

```python
data_products = requests.get(url)
data_products.status_code
```

```python
product_content = data_products.json()
product_content
```

```python
type(product_content)
```

```python
product_content.keys()
```

```python
extract_products = product_content.get('products')
```

```python
type(extract_products)
```

```python
extract_products[:3]
```

```python
sample_products = extract_products[:3]
type(sample_products)
```

```python
import pandas
```

```python
styled_products = pandas.DataFrame(sample_products)
styled_products
```

```python
styled_products.to_csv("sampleproducts.csv")
```

#### 2. Carts endpoint
